### Preprocessed ECoG 

```python
filtered = data.bandpass_filter(lowcut=0.2,highcut=60,fs = 100,inplace=True) 
```

```python
freq = np.array([i for i in range(10,150,15)]) 
scalo, motion,time = data.scalo(1,freq,start = 0,end = len(x)) # preprocess first 3000(30 sec) objects of data
```

Motrenko results

```python
n = len(time)
s = int(0.33*n/10)
X_train = X[:s]
X_test = X[-s:]
y_train = motion[:s]
y_test = motion[-s:] 
time_test = time[:s]
time_train = time[-s:]
```

```python
pls_motrenko_model = PLSRegression(n_components = 50)
pls_motrenko_model.fit(X_train,y_train)
y_pred = pls_motrenko_model.predict(X_test)
```

```python
plt.figure(figsize = (15,6))
plt.plot(time_test,y_pred[:,0], color = 'blue', alpha = 0.5)
plt.plot(time_test,y_train[:,0], color = 'green')
```

```python
from sklearn.metrics import r2_score
r2_score(y_test[:,0],y_pred[:, 0])
```

### Гипотезы нормального/гамма распределения 

```python
signal_data = np.array(x)
motion_data = np.array(y)
```

```python
#simple sync
time, signal_idx, motion_idx = np.intersect1d(signal_data[:,0], motion_data[:,0],assume_unique=True, return_indices=True)
ecog_signal = signal_data[signal_idx]
motion = motion_data[motion_idx]
```

```python
scaler = StandardScaler()
scaler.fit(ecog_signal)
ecog_signal_normed = scaler.transform(ecog_signal)
```

```python
N = len(time)
```

```python
%%time
m = get_mean(ecog_signal[:N], time[:N])
d = get_disp(m, ecog_signal[:N], time[:N])
```

```python
%%time
m_normed = get_mean(ecog_signal_normed[:N], time[:N])
d_normed = get_disp(m, ecog_signal_normed[:N], time[:N])
```

```python
%%time
#Calcuate bivariate normal distribution for each moment of time

mv = []
for i in range(len(m)):
    mv.append(multivariate_normal([m[i, 0], m[i, 1]], [[d[i,0], 0], [0, d[i, 1]]]))
```

```python
%%time
mv_normed = []
for i in range(len(m_normed)):
    mv_normed.append(multivariate_normal([m_normed[i, 0], m_normed[i, 1]], [[d_normed[i,0], 0], [0, d_normed[i, 1]]]))
```

```python
%%time
intens = get_intens(m, d, mv, ecog_signal[:N], time[:N])
```

```python
intens_normed = get_intens(m_normed, d_normed, mv_normed, ecog_signal[:N], time[:N])
```

```python
len(intens_normed)
```

```python
X = np.c_[m_normed, d_normed, intens_normed]
```

```python
from sklearn.linear_model import Ridge
```

```python
X.shape
```

```python
N = 70000
model = Ridge()
model.fit(X[:N], motion[:N, 1])

Y_pred = model.predict(X[N:])
```

```python
plt.figure(figsize = (15,6))
plt.plot(time[N:],Y_pred[:], color = 'green')
plt.plot(time[N:],motion[N:,1])
```

```python
r2_score(motion[N:,0], Y_pred)
```

### Линейная модель по параметрам среднего, дисперсии и интенсивности

```python
from sklearn.linear_model import LinearRegression
```

```python
n_sample = int(len(signal_data)/3)
```

```python
channel = 14
y_train = signal_data[:n_sample, channel]
```

```python
m1 = get_mean(signal_data[:, 1:], signal_data[:, 0])
d1 = get_disp(m1, signal_data[:, 1:], signal_data[:, 0])
```

```python
len(mv1)
```

```python
mv1 = []
for i in range(len(m1)-1):
    mv1.append(multivariate_normal([m1[i, 0], m1[i, 1]], [[d1[i,0], 0], [0, d1[i, 1]]]))
```

```python
intens1 = get_intens(m1[:1040000], d1[:1040000], mv1[:1040000], signal_data[:1040000, 1:], signal_data[:1040000, 0])
```

```python
x_1 = np.c_[m1[:1040000], d1[:1040000], intens1[:1040000]]
```

```python
reg = LinearRegression().fit(x_1[:n_sample], y_train)
```

```python
yp = reg.predict(x_1[n_sample:])
reg.coef_
```

```python
#plt.plot(range(20000, 40000),yp[20000:40000], alpha = 0.5)
#plt.plot(range(20000, 40000),y_train[20000:40000], alpha = 0.5)
plt.plot(range(len(yp)),yp, alpha = 0.5, color = 'blue')
plt.plot(range(len(yp)),signal_data[n_sample:1040000, channel], alpha = 0.5, color = 'green')
```

```python
mean_absolute_error(signal_data[n_sample:1040000, channel], yp)
```

```python
from sklearn.metrics import r2_score
r2_score(signal_data[n_sample:1040000, channel],yp)
```
