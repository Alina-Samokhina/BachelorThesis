```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.cross_decomposition import PLSRegression
import ECoG
```

```python
x1,y1 = ECoG.read_ECoG_from_csv("ecogs.csv","motions.csv")

data1 = ECoG.ECoG(x1,y1,downsample = True)
filtered1 = data1.bandpass_filter(lowcut=0.2,highcut=60,fs = 100,inplace=True) 

#works for a long time, needs about 1 hour to preprocess 15 min experiment
freq = np.array([i for i in range(10,150,15)]) 
scalo1, motion1,time1 = data1.scalo(1,freq,start = 0,end = 3000) # preprocess first 3000(30 sec) objects of data

X1 = scalo1.reshape((scalo1.shape[0],scalo1.shape[1]*scalo1.shape[2]*scalo1.shape[3]))

model1 = PLSRegression(n_components = 50)
model1.fit(X1,motion1)
Y_pred1 = model1.predict(X1)
```

```python
plt.figure(figsize = (15,6))
plt.plot(time1,Y_pred1[:,0], color = 'blue', label = 'predicted')
plt.plot(time1,motion1[:,0], color = 'orange', label = 'true')
plt.legend()
```

```python
x,y = ECoG.read_ECoG_from_csv("ECoG.csv","Motion.csv")

data = ECoG.ECoG(x,y,downsample = True)
filtered = data.bandpass_filter(lowcut=0.2,highcut=60,fs = 100,inplace=True) 

#works for a long time, needs about 1 hour to preprocess 15 min experiment
freq = np.array([i for i in range(10,150,15)]) 
scalo, motion,time = data.scalo(1,freq,start = 0,end = 3000) # preprocess first 3000(30 sec) objects of data

X = scalo.reshape((scalo.shape[0],scalo.shape[1]*scalo.shape[2]*scalo.shape[3]))
```

```python
model = PLSRegression(n_components = 50)
model.fit(X,motion)
Y_pred = model.predict(X)
```

```python
from sklearn.metrics import r2_score
r2_score(motion[:,0],Y_pred[:, 0])
```

```python
plt.figure(figsize = (15,6))
plt.plot(time,Y_pred[:,0])
plt.plot(time,motion[:,0])
```
